```python
def allUniqueTriplets(numProc):
    triplets = []
    for i in np.arange(0, numProc, 1):
        for j in np.arange(i, numProc, 1):
            for k in np.arange(j, numProc, 1):
                triplets.append((i,j,k))
    return triplets

def allUniquePairs(numProc):
    pairs = []
    for i in np.arange(0, numProc, 1):
        for j in np.arange(i, numProc, 1):
            pairs.append((i,j))
    return pairs

def containsTriplet(triple, triplets):
    for t in triplets:
        t0 = (t[0], t[1], t[2])
        t1 = (t[2], t[0], t[1])
        t2 = (t[1], t[2], t[0])
        t3 = (t[0], t[2], t[1])
        t4 = (t[1], t[0], t[2])
        t5 = (t[2], t[1], t[0])
        if t0 == triple or t1 == triple or t2 == triple or t3 == triple or t4 == triple or t5 == triple:
            return True
    return False

def containsPair(pair, pairs):
    for p in pairs:
        p0 = (p[0], p[1])
        p1 = (p[1], p[0])
        if p0 == pair or p1 == pair:
            return True
    return False

def containsTripletOnce(triple, triplets):
    count = 0
    for t in triplets:
        t0 = (t[0], t[1], t[2])
        t1 = (t[2], t[0], t[1])
        t2 = (t[1], t[2], t[0])
        t3 = (t[0], t[2], t[1])
        t4 = (t[1], t[0], t[2])
        t5 = (t[2], t[1], t[0])
        if t0 == triple or t1 == triple or t2 == triple or t3 == triple or t4 == triple or t5 == triple:
            count = count + 1
    return count

def containsPairOnce(pair, pairs):
    count = 0
    for p in pairs:
        p0 = (p[0], p[1])
        p1 = (p[1], p[0])
        if p0 == pair or p1 == pair:
            count = count + 1
    return count

# print(allUniqueTriplets(3))
# print(containsTripletOnce((0,0,2), allUniqueTriplets(3)))

def appendHelper(triplet, buf):
    if(len(buf) == 0):
        buf.append([triplet])
    else:
        buf[-1].append(triplet)
```

```python
def auta2B(numProc, proc):
    result = []
    i = 2
    b0Owner = proc
    b1Owner = proc
    b2Owner = proc
    rounds = np.arange(numProc, 0, -3).tolist()
    #print ("s = " + str(rounds))
    counter = 0
    numSubSteps = int(comb((numProc + 2 - 1), 2) / numProc)
    #print("numSubSteps: " + str(numSubSteps))
    for r, s in enumerate(rounds):
        #if((counter < (numSubSteps - 1)) or ((numProc % 2 == 0) and (counter < numSubSteps))):
        if(counter >= numSubSteps):
            continue
        #print("round " + str(r))
        for idx, t in enumerate(np.arange(s)):
            if(counter >= numSubSteps):
                continue
            if(idx != 0 or s != numProc):
                if(i == 0):
                    b0Owner = (b0Owner - 1) % numProc
                    #print ("shift right b0, new owner is " + str(b0Owner))
                if(i == 1):
                    b1Owner = (b1Owner - 1) % numProc
                    #print ("shift right b1, new owner is " + str(b1Owner))
                if(i == 2):
                    b2Owner = (b2Owner - 1) % numProc
                    #print ("shift right b2, new owner is " + str(b2Owner))
            #print("(" + str(b0Owner) + ", " + str(b1Owner) + ", " + str(b2Owner) + ")")
            result.append("(" + str(b0Owner) + ", " + str(b2Owner) + ")")
            counter = counter + 1
        if(counter < numSubSteps):    
            i = (i + 1) % 3
    if(numProc % 2 == 0):
        if(i == 0):
            b0Owner = (b0Owner - 1) % numProc
            #print ("shift right b0, new owner is " + str(b0Owner))
        if(i == 1):
            b1Owner = (b1Owner - 1) % numProc
            #print ("shift right b1, new owner is " + str(b1Owner))
        if(i == 2):
            b2Owner = (b2Owner - 1) % numProc
            #print ("shift right b2, new owner is " + str(b2Owner))
        #print("(" + str(b0Owner) + ", " + str(b1Owner) + ", " + str(b2Owner) + ")")
        result.append("(" + str(b0Owner)+ ", " + str(b2Owner) + ")")
    return result

def calcAuta2B(numProc):
    output = []
    for x in np.arange(numProc):
        output.append(auta2B(numProc, x))
    output = map(list, zip(*output))
    return output
```

```python
def auta2BAnd3B(numProc, proc, isRespaStep):
    result = []
    i = 2
    b0Owner = proc
    b1Owner = proc
    b2Owner = proc
    rounds = np.arange(numProc, 0, -3).tolist()
    counter = 0
    numSubSteps = int(comb((numProc + 2 - 1), 2) / numProc)
    for r, s in enumerate(rounds):
        for idx, t in enumerate(np.arange(s)):
            if((not isRespaStep) and (counter >= numSubSteps)):
                break
            if(idx != 0 or s != numProc):
                if(i == 0):
                    b0Owner = (b0Owner - 1) % numProc
                if(i == 1):
                    b1Owner = (b1Owner - 1) % numProc
                if(i == 2):
                    b2Owner = (b2Owner - 1) % numProc
            if(isRespaStep):
                result.append("(" + str(b0Owner) + ", " + str(b1Owner) + ", " + str(b2Owner) + ")")
            if(counter < numSubSteps):
                if(isRespaStep):
                    result[-1] = result[-1] + ", (" + str(b0Owner) + ", " + str(b2Owner) + ")"
                else:
                    result.append("(" + str(b0Owner) + ", " + str(b2Owner) + ")")
            if((counter == numSubSteps) and (numProc % 2 == 0)):
                if(isRespaStep):
                    result[-1] = result[-1] + ", (" + str(b0Owner) + ", " + str(b2Owner) + ")"
                else:
                    result.append("(" + str(b0Owner) + ", " + str(b2Owner) + ")")
            counter = counter + 1
        if((not isRespaStep) and (counter >= numSubSteps)):
                break
        i = (i + 1) % 3
    if((isRespaStep and ((numProc % 3) == 0)) or ((not isRespaStep) and ((numProc % 2) == 0))):
        if(i == 0):
            b0Owner = (b0Owner - 1) % numProc
        if(i == 1):
            b1Owner = (b1Owner - 1) % numProc
        if(i == 2):
            b2Owner = (b2Owner - 1) % numProc
        if(isRespaStep):
            result.append("(" + str(b0Owner) + ", " + str(b1Owner) + ", " + str(b2Owner) + ")")
        else:
            result.append("(" + str(b0Owner) + ", " + str(b2Owner) + ")")
    return result

def calcAuta2B3B(numProc, isRespa):
    output = []
    for x in np.arange(numProc):
        output.append(auta2BAnd3B(numProc, x, isRespa))
    output = map(list, zip(*output))
    return output
```

```python
def embedded2BAnd3BAlt(numProc, proc, isRespaStep):
    triplets = []
    pairs = []
    i = 0
    b0Owner = proc
    b1Owner = proc
    b2Owner = proc

    if(numProc == 1):
        # no shifts needed
        if(isRespaStep):
            appendHelper((b0Owner, b0Owner, b0Owner), triplets)

        appendHelper((b0Owner, b0Owner), pairs)
        return (triplets, pairs)

    if(numProc == 2):
        # only shift b2
        b2Owner = (proc + 1) % numProc

        if(isRespaStep):
            appendHelper((b1Owner, b1Owner, b1Owner), triplets)
            appendHelper((b1Owner, b1Owner, b2Owner), triplets)

        appendHelper((b1Owner, b1Owner), pairs)
        appendHelper((b1Owner, b2Owner), pairs)

        return (triplets, pairs)

    if(numProc == 3):
        # shift b0 and b2
        b0Owner = (proc - 1) % numProc
        b2Owner = (proc + 1) % numProc

        if(isRespaStep):
            appendHelper((b1Owner, b1Owner, b1Owner), triplets)
            appendHelper((b1Owner, b1Owner, b2Owner), triplets)
            appendHelper((b0Owner, b0Owner, b2Owner), triplets)
            appendHelper((b0Owner, b1Owner, b2Owner), triplets)

        appendHelper((b1Owner, b1Owner), pairs)
        appendHelper((b0Owner, b2Owner), pairs)
        return (triplets, pairs)
    
    b0Owner = (proc - 1) % numProc
    b1Owner = proc
    b2Owner = (proc + 1) % numProc

    counter = 0
    numSubSteps = int(comb((numProc + 2 - 1), 2) / numProc)

    rounds = np.arange(numProc - 3, 0, -3).tolist()
    for s in rounds:
        for idx, t in enumerate(np.arange(s)):
            # in the very first step (substep 1 of phase 1) we don't have to shift (right buffers are already there) 
            if(idx != 0 or s != numProc - 3):
                if(i == 0):
                    b0Owner = (b0Owner - 1) % numProc
                if(i == 1):
                    b1Owner = (b1Owner - 1) % numProc
                if(i == 2):
                    b2Owner = (b2Owner - 1) % numProc
                triplets.append([])
                if((counter - 1) < numSubSteps):
                    pairs.append([])
            else:
                # in the very first step (substep 1 of phase 1) we incorporate some 3-body and 2-body interactions we can calculate from the buffers we already have
                if(isRespaStep):
                    appendHelper((b1Owner, b1Owner, b1Owner), triplets)
                    appendHelper((b1Owner, b1Owner, b2Owner), triplets)
                    appendHelper((b0Owner, b0Owner, b2Owner), triplets)

                    appendHelper((b1Owner, b1Owner), pairs)
                    appendHelper((b0Owner, b1Owner), pairs)
                    counter = counter + 2
            if(s == numProc - 3):
                # See Embedded AUT Algorithm paper: In the first phase we incorporate also some additional 3-body calculations after buffer 0 is shifted
                if(isRespaStep):
                    appendHelper((b0Owner, b1Owner, b1Owner), triplets)

            if(isRespaStep):
                appendHelper((b0Owner, b1Owner, b2Owner), triplets)

            # calculate pair interactions until we have all necessary pairs (note: if the number of processors is divisible by 2, we need one extra round where interactions are distributed over processors)
            if((counter < numSubSteps or (numProc % 2 == 0 and counter == numSubSteps))):
                appendHelper((b0Owner, b2Owner), pairs)
                counter = counter + 1

            # since two-body interactions are fully embedded into the first phase of the algorithm we can stop the algorithm here, if we have a non-respa iteration
            if((not isRespaStep) and (counter >= numSubSteps)):
                break
        i = (i + 1) % 3
    # special case for threebody interactions when the number od processors is divisible by three
    if(numProc % 3 == 0 and isRespaStep):
        if(numProc > 3):
            if(i == 0):
                b0Owner = (b0Owner - 1) % numProc
            if(i == 1):
                b1Owner = (b1Owner - 1) % numProc
            if(i == 2):
                b2Owner = (b2Owner - 1) % numProc
        triplets.append([])
        appendHelper((b0Owner, b1Owner, b2Owner), triplets)
    return (triplets, pairs)

def calcEmbedded2B3BAlt(numProc, isRespa, threeBody):
    output = []
    for x in np.arange(numProc):
        calc = embedded2BAnd3BAlt(numProc, x, isRespa)
        if(threeBody):
            output.append(calc[0])
        else: 
            output.append(calc[1])
    output = map(list, zip(*output))
    return output
```

```python
def Simulation(numProc, respaStep, steps, threeBody):
    outputs = []
    for idx, x in enumerate(np.arange(steps)):
        if(idx % respaStep == 0):
            outputs.append(calcEmbedded2B3BAlt(numProc, True, threeBody))
        else:
            outputs.append(calcEmbedded2B3BAlt(numProc, False, threeBody))
    return outputs

numproc = 8
respaStep = 1
iterations = 1
simOutput1_3B = Simulation(numproc, respaStep, iterations, True)
simOutput1_2B = Simulation(numproc, respaStep, iterations, False)

for out in  simOutput1_3B:
    df = pd.DataFrame(out)
    with pd.option_context('display.max_rows', None, 'display.max_columns', None,'display.max_colwidth', -1):
        display(df)

for out in  simOutput1_2B:
    df = pd.DataFrame(out)
    with pd.option_context('display.max_rows', None, 'display.max_columns', None,'display.max_colwidth', -1):
        display(df)

simOutput2_3B = Simulation(numproc, respaStep, iterations, True)
simOutput2_2B = Simulation(numproc, respaStep, iterations, False)
#print(list(simOutput[0]))
# check for unique triplets
aut = allUniqueTriplets(numproc)
print("allUniqueTriplets: size: " + str(len(aut)) + ", elements: " + str(aut))
for idx, iteration in enumerate(simOutput2_3B):
    flatTriplets = []
    lstIteration = list(iteration)
    #print(tmp)
    #print(len(tmp))
    for substepIdx, substep in enumerate(lstIteration):
        #print(substepIdx)
        for rank, procData in enumerate(substep):
            #print("rank: " + str(rank) + str(procData))
            if(((numproc % 3) == 0) and (numproc > 3) and (rank >= (numproc // 3)) and substepIdx == len(list(lstIteration)) - 1):
                #print("skip")
                break
            if((numproc == 3) and (rank >= (numproc // 3)) and substepIdx == len(list(lstIteration)) - 1):
                flatTriplets.extend(procData[0:3])
                continue
            flatTriplets.extend(procData)
    print("flatTriplets: size: " + str(len(flatTriplets)) + ", elements: " + str(flatTriplets))
    allContainedOnce = []
    for t in aut:
        #print("check " + str(t) + " and " + str(flatTriplets))
        res = containsTripletOnce(t, flatTriplets)
        if(res == 1):
            allContainedOnce.append(True)
        else:
            allContainedOnce.append(False)
            print("Triplet " + str(t) + " is contained " + str(res) + " times")
    #print(allContainedOnce)
    print("all triplets occur exactly once in iteration " + str(idx) + ": " + str(all(allContainedOnce)))

# check for unique pairs
aup = allUniquePairs(numproc)
print("allUniquePairs: size: " + str(len(aup)) + ", elements: " + str(aup))
for idx, iteration in enumerate(simOutput2_2B):
    flatPairs = []
    lstIteration = list(iteration)
    #print(tmp)
    #print(len(tmp))
    for substepIdx, substep in enumerate(lstIteration):
        #print(substepIdx)
        for rank, procData in enumerate(substep):
            #print("rank: " + str(rank) + str(procData))
            if(((numproc % 2) == 0) and (rank >= (numproc // 2)) and substepIdx == len(list(lstIteration)) - 1):
                if(len(procData) == 1):
                    break
                else:
                    flatPairs.extend(procData[0:-1])
                continue
            flatPairs.extend(procData)
    print("flatPairs: size: " + str(len(flatPairs)) + ", elements: " + str(flatPairs))
    allContainedOnce = []
    for p in aup:
        #print("check " + str(t) + " and " + str(flatTriplets))
        res = containsPairOnce(p, flatPairs)
        if(res == 1):
            allContainedOnce.append(True)
        else:
            allContainedOnce.append(False)
            print("Pair " + str(p) + " is contained " + str(res) + " times")
    #print(allContainedOnce)
    print("all pairs occur exactly once in iteration " + str(idx) + ": " + str(all(allContainedOnce)))
```
